# Get data of all countries

```python
country_data = wb.economy.DataFrame()

```

```python
country_ids = country_data.index.to_list()
```

```python
country_ids
```

```python
# pull all data and join  
# define columns to join on 
join_columns = ['Country', 'Date']

# get life expectency
data_life_exp = get_data('SP.DYN.LE00.IN', country_ids)
data_life_exp.rename(columns={'Value': 'life_exp'}, inplace=True)

data_elect_acess = get_data('EG.ELC.ACCS.RU.ZS', country_ids)
data_elect_acess.rename(columns={'Value': 'access_to_electricity'}, inplace=True)
df = pd.merge(data_life_exp, data_elect_acess, on=join_columns, how='left')


data_trade_in_services = get_data('BG.GSR.NFSV.GD.ZS', country_ids)
data_trade_in_services.rename(columns={'Value': 'data_trade_in_services'}, inplace=True)
df = pd.merge(df, data_trade_in_services, on=join_columns, how='left')

data_foreign_investm= get_data('BM.KLT.DINV.WD.GD.ZS', country_ids)
data_foreign_investm.rename(columns={'Value': 'data_foreign_investm'}, inplace=True)
df = pd.merge(df, data_foreign_investm, on=join_columns, how='left')


data_stocks_traded = get_data('CM.MKT.TRAD.GD.ZS', country_ids)
data_stocks_traded.rename(columns={'Value': 'data_stocks_traded'}, inplace=True)
df = pd.merge(df, data_stocks_traded, on=join_columns, how='left')

data_companies_marketcap = get_data('CM.MKT.LCAP.GD.ZS', country_ids)
data_companies_marketcap.rename(columns={'Value': 'data_companies_marketcap'}, inplace=True)
df = pd.merge(df, data_companies_marketcap, on=join_columns, how='left')

data_gdp_per_energy_use = get_data('EG.GDP.PUSE.KO.PP.KD', country_ids)
data_gdp_per_energy_use.rename(columns={'Value': 'data_gdp_per_energy_use'}, inplace=True)
df = pd.merge(df, data_gdp_per_energy_use, on=join_columns, how='left')

data_alcohol_cons = get_data('SH.ALC.PCAP.LI', country_ids)
data_alcohol_cons.rename(columns={'Value': 'data_alcohol_cons'}, inplace=True)
df = pd.merge(df, data_alcohol_cons, on=join_columns, how='left')

data_perc_overweigth = get_data('SH.STA.OWAD.MA.ZS', country_ids)
data_perc_overweigth.rename(columns={'Value': 'data_perc_overweigth'}, inplace=True)
df = pd.merge(df, data_perc_overweigth, on=join_columns, how='left')

data_child_immunization = get_data('SH.IMM.IBCG', country_ids)
data_child_immunization.rename(columns={'Value': 'data_child_immunization'}, inplace=True)
df = pd.merge(df, data_child_immunization, on=join_columns, how='left')
```
